```python
# SIMULATE RAW LOGS STREAMING

import os
from pathlib import Path
import requests
import time
from datetime import datetime

PORT = 8000

END_POINT = 'logs/ingest'


url = f'http://localhost:{PORT}/{END_POINT}'


def sent_request(data):

    response = requests.post(url, json = data)

    if response == 200:
        print(response.json(), response.text)
    else:
        print(response)


script_dir = Path(os.getcwd())
project_dir = script_dir.parent

# main folder path

file_name = 'deanonymized_server_backup.log'
file_path = project_dir.joinpath(rf'test_logs\{file_name}')


with open(file_path) as my_file:

    previous_time = None

    for i, line in enumerate(my_file):

        # i==100 - contains 'WARN'
        # i >= 230 and i<= 250 - contains invalid logs

        if i <= 40:

            cleared_log = line.rstrip('\n')

            timestamp_str = line.split(']')[0].strip('[]')

            try:
                current_time = datetime.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S,%f')
            except ValueError:
                previous_time = None
                # if it some wierd log without date just sent it over


            # Sleep to simulate delay
            if previous_time is not None:
                delta = (current_time - previous_time).total_seconds()
                time.sleep(delta)

            print(cleared_log)  # or process the log as needed

            previous_time = current_time

            sent_request(cleared_log)




```

```python
from pydantic import BaseModel
import json

# Define the model
class User(BaseModel):
    id: int
    name: str
    email: str

# Create a model instance
user = User(id=1, name="Alice", email="alice@example.com")

# Convert to JSON (as a string)
user_json = user.model_dump_json()
print("Serialized to JSON string:")
print(user_json)

# Deserialize back into a model
user_obj = User.model_validate_json(user_json)
print("\nDeserialized back to Pydantic model:")
print(user_obj)
```
